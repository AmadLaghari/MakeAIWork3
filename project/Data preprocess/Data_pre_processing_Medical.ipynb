```python
import pandas as pd

from google.colab import drive
drive.mount('/content/gdrive')
```

```python
data= "/content/gdrive/My Drive/Datasets/Medische_dataset/medisch-centrum-randstad.csv"

df = pd.read_csv(data)
```

```python
df
```

```python
# Convert length to meters
df["length"] = df["length"] / 100

# Calculate the BMI
df["BMI2"] = df["mass"] / (df["length"] ** 2)
df["BMI2"] = df["BMI2"].round(1)
```

```python
# Save the BMI column
bmi_column = df["BMI2"]

# Insert the BMI column at the second position
df.insert(1, "BMI", bmi_column)
```

```python
df = df.drop(columns=["BMI2"], axis=1)
df = df.drop(columns=["length"], axis=1)
df = df.drop(columns=["mass"], axis=1)
```

```python
df
```

```python
data_extra= "/content/gdrive/My Drive/Datasets/Medische_dataset/medisch-centrum-randstad-extra.csv"
df2 = pd.read_csv(data_extra)
```

```python
df2
```

```python
# Convert length to meters
df2["length"] = df2["length"] / 100

# Calculate the BMI
df2["BMI2"] = df2["mass"] / (df2["length"] ** 2)
df2["BMI2"] = df2["BMI2"].round(1)

# Save the BMI column
bmi_column = df2["BMI2"]

# Insert the BMI column at the second position
df2.insert(1, "BMI", bmi_column)
```

```python
df2 = df2.drop(columns=["BMI2"], axis=1)
df2 = df2.drop(columns=["length"], axis=1)
df2 = df2.drop(columns=["mass"], axis=1)
```

```python
df2
```

```python
df3 = pd.concat([df, df2], axis=0)
```

```python
df3
```

```python
df3_missing_rows = df3[df3.isnull().any(axis=1)]
print(df3_missing_rows)
```
